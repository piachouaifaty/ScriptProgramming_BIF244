```python
#demo of bootstrapping via resampling with replacement
import random
MyReps = 50 #number of resamplings to conduct

#for the demo, create an arbitrary list of numbers from 0 to 19
#normally this would be the data
#indeces of rows in dataframe

MyDataList = range(20)

# loop repeatedly to create resampled lists:
# if Samples is not specified, takes it to be 10 by defaults

#DataList usually would include just the indeces of my data observations

def bootstrap(DataList, Samples=10):
    Bootstraps=[] #List that will store the random lists
    random.seed(1) #to ensure reproducibility
    
    for x in range(Samples):
        Resample = [random.choice(DataList) for Y in DataList]
        Bootstraps.append(Resample)
    return Bootstraps

Bootstraps = bootstrap(MyDataList,MyReps)
for Z in Bootstraps:
    print(Z)
```
