```python
#2. Write a python script that takes as input two tab delimited text files and merge them together. 
#The two files should have one common column based on which you will merge them. (names.txt and majors.txt)

#a. You should ask the user to enter the path to the first file and check if it exists and readable. 
#If not, ask him to reenter it.

#b. Ask the user to enter the path to the second file and also check if it exists or not.

#c. Ask the user to enter the column title that needs to be used to merge the files
#(examples: Student_ID)

#d. Save the column names (header) of the first file in a list

#e. Save the column names (header) of the second file in a second list

#f. Check if the column name specified in part c. (Student_ID) is found in both lists. 
#If not print out on the terminal that the column is not found and specify in which file it
#is not found.

#g. If the column is found in both files, create two dictionaries dict1 and dict2, one for 
#each file, with the key being the value of the “Student_ID” (201909199) and the
#value the whole remaining columns for that row.

#h. Create a list, allKeys, that include all the Keys in both dictionaries combines, keys of
#dict1 and keys of dict2. Careful!!! Not all IDs are in both files!!!

#i. Create a new dictionary, mergedDict, with the keys being allKeys created in part h.
#and the values the concatenation of the value in dict1 and dict2.

#j. Print out the mergedDict and write it to a file called “mergedFiles.txt”
#k. Count the number of Ids in that are common to both files and the number unique in
#each file.
#l. Ask the user to enter the ID of one student and print out all the value of that ID in mergedDict.
```

```python
import os
import sys
import pandas as pd

#a

files_in_dir=os.listdir()
print("Please enter file1")
file1=input()
if file1=="" or not file1 in files_in_dir:
    sys.stderr.write("Error, no file entered/no such file in directory, please enter a file name for file1")
    file1=input()

#b

print("Please enter file2")
file2=input()
if file2=="" or not file2 in files_in_dir:
    sys.stderr.write("Error, no file entered/no such file in directory, please enter a file name for file2")
    file2=input()
    
#c
print("Please enter column to merge")
col2merge=input()

f1=open(file1)
f2=open(file2)

f1string=f1.read()
f2string=f2.read()

#saved headers as strings
f1head=f1string.split('\n')[0]
f2head=f2string.split('\n')[0]

#split header strings
#d-e
f1headlist=f1head.split()
f2headlist=f2head.split()

#print(f1headlist)
#f
foundf1 = False
foundf2 = False

if col2merge in f1headlist:
    foundf1=True   
    
if col2merge in f2headlist:
    foundf2=True

if foundf1==False:
    print(col2merge+ " not found in " + file1)

if foundf2==False:
    print(col2merge+ " not found in " + file2)

if foundf1 and foundf2:
    dict1={}
    dict2={}

#g
f1stringlines=f1string.split('\n')  
count=1
while count < len(f1stringlines):
    s=f1stringlines[count]
    s=s.split('\t')
    key=s[0]
    val=[]
    for i in range(1,len(s)):
        val.append(s[i])
    dict1[key]=val
    count=count+1
#print(dict1)

f2stringlines=f2string.split('\n')  
count=1
while count < len(f2stringlines):
    s=f2stringlines[count]
    s=s.split('\t')
    key=s[0]
    val=[]
    for i in range(1,len(s)):
        val.append(s[i])
    dict2[key]=val
    count=count+1
#print(dict2)

listid1=list(dict1.keys())
listid2=list(dict2.keys())
#print(listid1, listid2)

#k
setid1=set(listid1)
setid2=set(listid2)

print("common ids: "+str(len(setid1.union(setid2))-len(setid1.intersection(setid2))))

count=0
for id in setid1:
    if not id in setid2:
        count=count+1
print("Ids only in " +file1+": "+str(count))

count=0
for id in setid2:
    if not id in setid1:
        count=count+1
print("Ids only in " +file2+": "+str(count))



#h
listid=listid1+listid2
#print("not unique: ")
#print(listid)

#print("\n unique: ")
idset=set(listid)
#print(idset)

#i
mergedDict={}
for id in idset:
    key=id
    if key in dict1 and key in dict2:
        val=dict1[key]+dict2[key]
    elif key in dict1:
        val=dict1[key]
    else:
        val=dict2[key]
    mergedDict[key]=val

print(mergedDict)

#l
print("Enter ID to get info: ")
id=input()
print("\nInfo:\n")
print(mergedDict[id])

outfilename = "mergedFiles.txt"
writeout = True

if writeout:
    # Open the output file
    outfile = open(outfilename, 'w')
    outfile.write(mergedDict + '\n')
    outfile.close()
```
